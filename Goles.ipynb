# Estadísticas liga chilena

```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

dataframes_jugadores = {}
paginas = ['goles', 'goles-encajados', 'disparos', 'regates', 'pases-buenos', 'paradas-del-portero',
          'partidos-jugados-de-titular', 'minutos-totales-disputados', 'total-de-sustituciones-en-las-que-sale-del-campo',
          'total-de-sustituciones-en-las-que-entra-al-campo',
          'porcentaje-penaltis-parados-por-el-portero', 'intervenciones', 'asistencias-de-gol', 'asistencias-totales',
          'regates-satisfactorios', 'regates-en-los-que-pierdes-el-balon', 'porcentaje-de-regates-satisfactorio',
          'partidos-disputados', 'balones-disputados-donde-se-ha-perdido-la-posesion', 'balones-disputados-donde-se-ha-ganado-la-posesion',
          'balones-disputados', 'porcentaje-de-balones-disputados-donde-se-ha-ganado-la-posesion', 'duelos-uno-contra-uno-perdidos-sobre-la-posesion-del-balon-en-el-aire',
          'duelos-uno-contra-uno-ganados-sobre-la-posesion-del-balon-en-el-aire', 'duelos-uno-contra-uno-sobre-la-posesion-del-balon-en-el-aire',
          'faltas-cometidas', 'faltas-recibidas', 'goles-encajados-de-penalti', 'goles-en-propia-puerta',
          'goles-de-penalti', 'goles-de-falta', 'centros-fallados', 'centros-buenos', 'pases', 'fueras-de-juego',
          'tarjetas-rojas', 'tarjetas-amarillas', 'disparos-a-puerta', 'disparos-de-falta', 'disparos-al-poste',
          'entradas-que-terminan-en-falta', '']
```

```python
for pagina in paginas:
    
    goles = requests.get('https://chile.as.com/resultados/futbol/chile/2021/ranking/jugadores/'+pagina+'/')
    content_goles = goles.text
    soup = BeautifulSoup(content_goles, 'html.parser')

    players = soup.find_all('span', attrs = {'class': 'name'})
    pos = soup.find_all('span', attrs = {'class': 'info'})
    team = soup.find_all('td', attrs = {'class': 'col-team'})
    est = soup.find_all('td', attrs = {'class': 'cantidad'})
    
    list_names = []
    list_team_name = []
    list_est = []
    list_position = []

    for i in range(len(players)):
        name = players[i].text
        team_name = team[i].text[3:-2]
        position = pos[i].text
        estadistica = int(''.join(filter(str.isdigit, est[i].text)))
        list_names.append(name)
        list_team_name.append(team_name)
        list_position.append(position)
        list_est.append(estadistica)

    dict_estadisticas = {'Player_name':list_names, 'Position': list_position, 'Team_name': list_team_name, pagina: list_est}
    dataframes_jugadores[pagina] = pd.DataFrame(dict_estadisticas)
```
